```python
def sheetToClausesList(sheet, necessary_clauses):
    ''' 
    Takes as input the copy of a simplified sentence spreadsheet,
    Outputs the list of the different clauses without duplicates.
    arguments : 1. the copy of a sheet, 2. a list of necessary clauses
    '''
    
    #separate in lines
    versions = sheet.split('\n')

    #create lists with the same number of elements as there was of cells
    liste = []
    for v in versions:
        liste.append(v.split('\t'))

    #adding a space in the empty cells from the first line of the spreadsheet
    for num, word in enumerate(liste[0]):
        if word == '':
            for i in range(len(liste)):
                liste[i][num] = '*'

    #creating a list where the sentence words of each phrase are joined
    joined = []
    for l in liste:
        joined.append(''.join(l).split('*'))
    clauses = []
    for num, clause in enumerate(joined[0]):
        temp_list = []
        for i in range(len(joined)):
            temp_list.append(joined[i][num])
        clauses.append(temp_list)
        temp_list = []
    
    #eliminating duplicates and empty elements
    for i in range(len(clauses)):
        clauses[i] = list(filter(None, list(set(clauses[i]))))
    
    # adding the empty strings everywhere but where the clause is necessary
    for i in range(len(clauses)):
        if i+1 not in necessary_clauses:
            clauses[i].append('')
    return clauses
```

```python
def minimal_sheet(clauses_list):
    '''
    Creates a minimal sheet from the list of clauses. 
    '''
    num = 0
    for clauses in clauses_list:
        if len(clauses) > num:
            num = len(clauses)

    sheet = ''
    for n in range(num):
        new = []
        for i in range(len(clauses_list)):
            if len(clauses_list[i]) > n:
                sheet += clauses_list[i][n]+'\t'
        sheet += '\n'
    return sheet
```

```python
def generateSentences(text, needed_clauses):
    '''
    generates the output of the minimal sheet and all the sentences below
    '''
    clauses_list = sheetToClausesList(text, needed_clauses)
    sentences = getSentencesAfter(clauses_list, 1)
    output = ''
    output += minimal_sheet(clauses_list)+''
    output += str(len(sentences))+'\n'
    for s in sentences:
        output += re.sub(r'\s+', r' ', s.strip())+'\n'
    return output
```
